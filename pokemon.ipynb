```python
# Important fer una còpia de seguretat de tot el df abans de continuar.
import copy
pokedex2 = copy.deepcopy(pokedex)

# Màscara de substitució de valors.

# Pas 1. Aplicar la màscara per seleccionar els que són de tipus elèctric.
mask_electrics = (pokedex2.loc[ : , 'Type 1' ] == 'Electric') | (pokedex2.loc[ : , 'Type 2' ] == 'Electric')


# Pas 2. Aplicar la màscara; és a dir, modificar només les files del dataframe que compleixen la màscara.
#  df. 🗲 => U+1F5F2
pokedex2.loc[mask_electrics,['Type 1'] ] = "🗲"
pokedex2.loc[mask_electrics,['Type 2'] ] = "🗲"

# Pas 3. Validació. Consultem el resultat després d'aplicar la màscara.
print(pokedex2.loc[mask_electrics, : ])

# Comentari, també converteix els que no tenen Type 2. Ara ho arreglarem.
```
